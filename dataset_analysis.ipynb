```python
import pandas as pd

res = pd.read_csv("results.csv", index_col=0)
res['ARI'] = res['ARI'].apply(lambda x: round(x, 2))
res['NMI'] = res['NMI'].apply(lambda x: round(x, 2))

B_minus = res[res['name'].apply(lambda x: x[:2] == 'B-')]
B_plus = res[res['name'].apply(lambda x: x[:2] == 'B+')]
```
