```python
import pandas as pd
import numpy as np

res = pd.read_csv("results.csv", index_col=0)
res = res[np.logical_not(res['dataset'].isin(['Fungi_genera_6', 'Fungi_genera_7']))]
res['ARI'] = res['ARI'].apply(lambda x: round(x, 2))
res['NMI'] = res['NMI'].apply(lambda x: round(x, 2))

B_minus = res[res['name'].apply(lambda x: x[:2] == 'B-')].reset_index(drop=True)
B_plus = res[res['name'].apply(lambda x: x[:2] == 'B+')].reset_index(drop=True)

cols_new_names = {'BSP_GAP_3': 'BSP_3',
                  'BSP_GAP_4': 'BSP_4',
                  'BSP_GAP_5': 'BSP_5',
                  'Fungi_genera_3': 'Fungi_3',
                  'Fungi_genera_4': 'Fungi_4',
                  'Fungi_genera_5': 'Fungi_5',
                  }
```

```python
from matplotlib.colors import ListedColormap

fungi_cols = ['Fungi_genera_3',	'Fungi_genera_4',	'Fungi_genera_5']
bsp_cols = ['BSP_GAP_3',	'BSP_GAP_4',	'BSP_GAP_5']

KM_ids = ['B+ A0 KM', 
          'B+ A1 KM', 
          'B+ A1 KM 3d']

IKM_ids = ['B+ A0 IKM', 
           'B+ A1 IKM', 
           'B+ A2 IKM', 
           'B+ A1 IKM 3d', 
           'B+ A2 IKM 3d',] 

W_diff_algs = [
  'B+ A0 IKM A1',
  'B+ A0 IKM A2',
  'B+ A0 IKM A1 3d', 
  'B+ A0 IKM A2 3d',
  
  'B+ A1 IKM A0', 
  'B+ A1 IKM A2', 
  'B+ A1 IKM A2 3d',
  
  'B+ A2 IKM A0', 
  'B+ A2 IKM A1', 
  'B+ A2 IKM A1 3d',
  
  'B+ A0 KM A1', 
  'B+ A0 KM A1 3d', 
  'B+ A1 KM A0', 
  'B+ A2 KM A0', 
  'B+ A2 KM A1', 
  'B+ A2 KM A1 3d', 
  ]

B_minus_algs = [
  'B- A0 IKM', 
  'B- A1 IKM', 
  'B- A2 IKM', 
  'B- A1 IKM 3d', 
  'B- A2 IKM 3d', 
  'B- A0 KM', 
  'B- A1 KM',
  'B- A1 KM 3d', 
  ]

algs_new_names = {
    'B+ A0 KM' : 'BKMn', 
    'B+ A1 KM' : 'BKMc', 
    'B+ A1 KM 3d' : 'BKMcs',
    
    'B+ A0 IKM' : 'BIKMn', 
    'B+ A1 IKM' : 'BIKMc', 
    'B+ A2 IKM' : 'BIKMo', 
    'B+ A1 IKM 3d' : 'BIKMcs', 
    'B+ A2 IKM 3d' : 'BIKMos',
    
    
    'B+ A0 IKM A1' : 'BnIKMc',
    'B+ A0 IKM A2' : 'BnIKMo',
    'B+ A0 IKM A1 3d' : 'BnIKMcs', 
    'B+ A0 IKM A2 3d' : 'BnIKMos',
    
    'B+ A1 IKM A0' : 'BcIKMn', 
    'B+ A1 IKM A2' : 'BcIKMo', 
    'B+ A1 IKM A2 3d' : 'BcIKMos',
    
    'B+ A2 IKM A0' : 'BoIKMn', 
    'B+ A2 IKM A1' : 'BoIKMc', 
    'B+ A2 IKM A1 3d' : 'BoIKMcs',
    
    'B+ A0 KM A1' : 'BnKMc', 
    'B+ A0 KM A1 3d' : 'BnKMcs', 
    'B+ A1 KM A0' : 'BcKMn', 
    'B+ A2 KM A0' : 'BoKMn', 
    'B+ A2 KM A1' : 'BoKMc', 
    'B+ A2 KM A1 3d' : 'BnKMcs',
    
    
    'B- A0 IKM' : 'IKMn', 
    'B- A1 IKM' : 'IKMc', 
    'B- A2 IKM' : 'IKMo', 
    'B- A1 IKM 3d' : 'IKMcs', 
    'B- A2 IKM 3d' : 'IKMos',
     
    'B- A0 KM' : 'KMn', 
    'B- A1 KM' : 'KMc',
    'B- A1 KM 3d' : 'KMcs',
}


def get_res_table(res_df, algs, sep=' / '):
  res_df_max = res_df[['name', 'dataset', 'ARI', 'NMI']].groupby(by=['name', 'dataset']).max().reset_index()
  format_func = lambda x: "{:1.2f}".format(x)[1:] if x < 1 else "{:1.2f}".format(x)
  res_df_max['ARI \ NMI'] = res_df_max['ARI'].apply(format_func) + sep + res_df_max['NMI'].apply(format_func)
  # ari_pivot = res_df_max.pivot(index='name', columns='dataset', values='ARI').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
  table_df = res_df_max.pivot(index='name', columns='dataset', values='ARI \ NMI').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
  # s = table_df.style

  # cmap = ListedColormap(["red", "darkorange", "gold", "lawngreen", 'forestgreen', "darkgreen"])

  # s = s.background_gradient(axis=None, cmap=cmap, low=0., high=0.,
  #                             gmap=ari_pivot.apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=0))
  return table_df.rename(index=algs_new_names)


def get_res_table_max_and_mean(res_df, algs = [], sep=' / ', var=1):
  res_df['ARI+NMI'] = 100*res_df['ARI'] + res_df['NMI']
  idx1 = (res_df[['name', 'dataset', 'ARI+NMI']].groupby(by=['name', 'dataset']).transform("max") == res_df[['ARI+NMI']]).all(axis=1)
  
  B_df = res_df[idx1].groupby(by=['name', 'dataset', 'ARI', 'NMI']).min().reset_index()
  

  B_df = B_df[['name', 'dataset', 'beta1', 'beta2']].set_index(['name', 'dataset', 'beta1', 'beta2']).join(res_df.set_index(['name', 'dataset', 'beta1', 'beta2']), on=['name', 'dataset', 'beta1', 'beta2']).reset_index()

  B_df1 = B_df[['name', 'dataset', 'ARI', 'NMI']].groupby(by=['name', 'dataset']).max().reset_index()
  B_df2 = B_df[['name', 'dataset', 'ARI', 'NMI']].groupby(by=['name', 'dataset']).mean().reset_index()
  B_df3 = B_df[['name', 'dataset', 'ARI', 'NMI']].groupby(by=['name', 'dataset']).std().reset_index()
  
  format_func = lambda x: "{:1.2f}".format(x)[1:] if x < 1 else "{:1.2f}".format(x)
  if var == 1:
    B_df1['ARI | NMI'] = B_df1['ARI'].apply(format_func) + sep + B_df1['NMI'].apply(format_func)
    s1 = B_df1.pivot(index='name', columns='dataset', values='ARI | NMI').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
    s1 = s1.rename(index=algs_new_names)
    
    B_df2['ARI | NMI'] = B_df2['ARI'].apply(format_func) + sep + B_df2['NMI'].apply(format_func)
    s2 = B_df2.pivot(index='name', columns='dataset', values='ARI | NMI').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
    s2 = s2.rename(index=algs_new_names)
  else:
    B_df1['ARI | ARI mean | ARI std'] = B_df1['ARI'].apply(format_func) + sep + B_df2['ARI'].apply(format_func) + sep + B_df3['ARI'].apply(format_func)
    s1 = B_df1.pivot(index='name', columns='dataset', values='ARI | ARI mean | ARI std').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
    s1 = s1.rename(index=algs_new_names)
    
    B_df2['NMI | NMI mean | NMI std'] = B_df1['NMI'].apply(format_func) + sep + B_df2['NMI'].apply(format_func) + sep + B_df3['NMI'].apply(format_func)
    s2 = B_df2.pivot(index='name', columns='dataset', values='NMI | NMI mean | NMI std').loc[algs, fungi_cols + bsp_cols].rename(columns=cols_new_names)
    s2 = s2.rename(index=algs_new_names)
  return s1, s2
```

```python
s1, s2 = get_res_table_max_and_mean(B_minus, B_minus_algs, var=2)
s1.to_csv("B_minus_ARI_max_and_mean.csv")
s2.to_csv("B_minus_NMI_max_and_mean.csv")
s1
```
